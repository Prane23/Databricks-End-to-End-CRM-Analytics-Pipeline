## lets initialize Catalog

```python
%sql
CREATE CATALOG IF NOT EXISTS salesdb
```

## Lets create necessary schemas (bronze, silver and gold) 

```python
%sql
USE CATALOG salesdb;
CREATE SCHEMA IF NOT EXISTS bronze;
CREATE SCHEMA IF NOT EXISTS silver;
CREATE SCHEMA IF NOT EXISTS gold;
```

## lets create volumne to drop raw data 


```python
%sql
CREATE VOLUME IF NOT EXISTS salesdb.bronze.sales_data_volume
COMMENT 'Volume for raw source files (CSV)';
```
